## 2. Membaca file DATA C-DESA NOBOREJO.csv

```python
# Membaca file DATA C-DESA NOBOREJO.csv
file_path = 'DATA C-DESA NOBOREJO.csv'
df = pd.read_csv(file_path)
```

```python
# Menampilkan 5 data teratas
df.head()
```

## 3. Preproses data untuk membuat membaca file .csv lebih lebih mudah

```python
# Menghapus baris-baris kosong
df = df.dropna(how='all')

# Menghapus kolom-kolom kosong
df = df.dropna(axis=1, how='all')


# Menampilkan 5 data teratas
df.head()
```

```python
# Menghapus baris pertama yang berisi "KELURAHAN NOBOREJO"
df = df[df.apply(lambda row: row.astype(str).str.contains('KELURAHAN NOBOREJO').any(), axis=1) == False]
```

```python
# Membuat header
df.columns = ['NO', 'NAMA PEMILIK', 'ALAMAT', 'C-DESA', 'LUAS', 'KET', 'REDUNDANT', 'DESKRIPSI']

# Mentampilkan 5 data teratas
df.head()
```

```python
# Identifikasi dan menghitung baris-baris duplikat
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
print(duplicate_rows)
```

```python
# Menghapus baris-baris duplikat
df = df.drop_duplicates()

# Mentampilkan 5 data teratas
df.head()
```

```python
# Menghapus baris pertama
df = df.iloc[1:]

# Mentampilkan 5 data teratas
df.head()
```

```python
# Menghapus kolom 'REDUNDANT'
df = df.drop(columns=['REDUNDANT'])

# Mentampilkan 5 data teratas
df.head()
```

```python
# Save ke file .csv
df.to_csv('DATA C-DESA NOBOREJO preprocessed.csv', index=False)
```
